```python
df = pd.read_csv('sample_mkt_data_final.csv')
df
```

```python
def classification(x):
    if x.find('google') != -1:
        return 'google'
    elif x.find('youtube') != -1:
        return 'youtube'
    elif x.find('yahoo') != -1:
        return 'yahoo'
    elif x.find('quora') != -1:
        return 'quora'
    elif x.find('facebook') != -1:
        return 'facebook'
    elif x.find('reddit') != -1:
        return 'reddit'
    elif x.find('qiita') != -1:
        return 'qiita'
    elif x.find('baidu') != -1:
        return 'baidu'
    elif x.find('sogou') != -1:
        return 'sogou'
    else:
        return x
```

# Biểu đồ thể hiện conversion rate của từng source

```python
source_conversion0 = df.groupby('source')[['transactions']].count()
source_visit0 = df.groupby('source')['visits'].count()
source_conversion0['visits'] = source_visit0
source_conversion0['conversion_rate'] = source_conversion0['transactions'] / source_conversion0['visits'] * 100
source_conversion0.sort_values(by = 'conversion_rate', ascending = False)[0:15]
```

# Combine conversion rate and bounce rate cua source

```python
combine_source = pd.merge(source_conversion0, source_bounce0, how='inner', left_on = 'source', right_on = 'source').sort_values(by = 'visits_x', ascending = False)
combine_source
```
